```python
from transformers import AutoModelForCausalLM, AutoTokenizer
# Load model and tokenizer
model = AutoModelForCausalLM.from_pretrained(
    "microsoft/Phi-3-mini-4k-instruct",
    device_map="cuda",
    torch_dtype="auto",
)
tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-3-mini-4k-instruct")
```

```python
print(model)
```

```python
from transformers import pipeline
# Create a pipeline
generator = pipeline(
"text-generation",
model=model,
tokenizer=tokenizer,
return_full_text=False,
max_new_tokens=500,
do_sample=False,
temperature = 0.5
)
```

```python
# The prompt (user input / query)
messages = [
{"role": "user", "content": "Create a funny joke about cricket."}
]
# Generate output
output = generator(messages)
print(output[0]["generated_text"])
```

```python
import pandas as pd
from urllib import request

# Download and parse playlist data
data = request.urlopen('https://storage.googleapis.com/maps-premium/dataset/yes_complete/train.txt')
lines = data.read().decode("utf-8").split('\n')[2:]  # Skip metadata
playlists = [s.rstrip().split() for s in lines if len(s.split()) > 1]

# Load song metadata (titles and artists)
songs_file = request.urlopen('https://storage.googleapis.com/maps-premium/dataset/yes_complete/song_hash.txt')
songs_file = songs_file.read().decode("utf-8").split('\n')
songs = [s.rstrip().split('\t') for s in songs_file]
songs_df = pd.DataFrame(data=songs, columns=['id', 'title', 'artist'])
songs_df = songs_df.set_index('id')
```

```python
print('Playlist #1:\n', playlists[0])
print('Playlist #2:\n', playlists[1])
```

```python
!pip install gensim
```

```python
from gensim.models import Word2Vec

# Train embeddings for songs based on playlist co-occurrence
model = Word2Vec(
    playlists,
    vector_size=32,        # Each song gets a 32-dimensional embedding
    window=20,             # Consider 20 songs on each side as context
    negative=50,           # Use 50 negative samples per positive example
    min_count=1,           # Include all songs
    workers=4              # Use multiple CPU cores
)
```

```python
import numpy as np
def print_recommendations(song_id):
    # Find the 5 most similar songs
    similar_songs = np.array(
        model.wv.most_similar(positive=str(song_id), topn=5)
    )[:, 0]
    return songs_df.iloc[similar_songs]

# Test with Michael Jackson's "Billie Jean" (song ID 3822)
print_recommendations(3822)
```

```python
prompt = "The capital of France is"

# Convert text to token IDs
input_ids = tokenizer(prompt, return_tensors="pt").input_ids.to("cuda")

# Get the output before the LM head
model_output = model.model(input_ids)

# Pass through the LM head to get token probabilities
lm_head_output = model.lm_head(model_output[0])

# Extract the last token's probability scores and find the highest
token_id = lm_head_output[0, -1].argmax(-1)
print(tokenizer.decode(token_id))
```
